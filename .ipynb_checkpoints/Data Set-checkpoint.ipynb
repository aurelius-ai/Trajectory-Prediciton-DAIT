```python
stanf_data = pd.read_csv('./data/train/stanford/hyang_4.txt', header = None,
                 names = ['frameNb','id', 'x','y'],delimiter=' ')
id_unique = np.unique(np.array(stanf_data['id']))
c = np.zeros(len(stanf_data))
#for i in range(len(a)):
#    b = stanf_data['id'][stanf_data['id']==a[i]]
#    c[b.index]=i
#
#c = np.asarray(c,dtype = int)
#stanf_data['id']=c
#
#stanf_data.sort_values(['id'], ascending=[True], inplace=True)
#stanf_data.index = range(len(stanf_data))
```

## la boucle va chercher dans les id des piétons (id_unique). Si tu enlèves le slicing (id_unique[5:9]) et que tu met tout l'array il faut que tu mettes la condition au dessus du plot pour éviter qu'il y ai trop de figures qui se fasse et que ca fasse buger ton ordi !

```python
av_disp_err = np.array([])
fin_disp_err = np.array([])
pd.options.mode.chained_assignment = None ## Enlève un warning qui nous conernent pas


for i in id_unique[5:9]:
    a = stanf_data[stanf_data['id']==i]
    a.index = range(len(a))
    
    # Shift to 0,0
    x_shift, y_shift = a['x'][0] ,a['y'][0]
    a['x'] = a['x']-x_shift
    a['y'] = a['y']-y_shift
    k = 1
    ## Rotation pour que ca parte vers le haut le premier segment
    while sum(a.loc[k,['x','y']]==0)==2 and k < len(a)-1:
        k += 1
    if k<len(a)-1:
        thet = np.arccos(a['y'][k]/(np.linalg.norm(a.loc[k,['x','y']])))
        if a['x'][k]<0:
            thet=-thet
        rot_mat = np.array([[np.cos(thet),-np.sin(thet)],[np.sin(thet),np.cos(thet)]])
        for j in range(len(a)):
            a.loc[j,['x','y']] = rot_mat.dot(a.loc[j,['x','y']])
    
    ## Separe en deux parties, pour la prediction
    mid = np.around(len(a)/2)
    a_past = a.loc[0:mid,:]
    a_pred = a.loc[mid:,:]
    
    # Prediction lineaire: Vitesse et direction constante
    if (a_past['x'][len(a_past)-1]-a_past['x'][len(a_past)-2])==0:
        x_lin = a_past['x'][len(a_past)-1]*np.ones(len(a_pred))
        diff = a_past['y'][len(a_past)-1]-a_past['y'][len(a_past)-2]
        y_lin = range(len(a_pred))*diff+a_past['y'][len(a_past)-1]
    else:
        slope = (a_past['y'][len(a_past)-1]-a_past['y'][len(a_past)-2])/(a_past['x'][len(a_past)-1]-a_past['x'][len(a_past)-2])
        ordor = a_past['y'][len(a_past)-1] - slope*a_past['x'][len(a_past)-1]
        diff = a_past['x'][len(a_past)-1]-a_past['x'][len(a_past)-2]
        x_lin = range(len(a_pred))*diff+a_past['x'][len(a_past)-1]
        y_lin = slope*x_lin+ordor
    
    ## MSE
    av_disp_err = np.append(av_disp_err,(sum((a_pred['y']-y_lin)**2+(a_pred['x']-x_lin)**2))/len(a))
    
    ## Final displa
    fin_disp_err = np.append(fin_disp_err,sum((a_pred['y'].tail(1)-y_lin[-1])**2+(a_pred['x'].tail(1)-x_lin[-1])**2))
    
    #if i%5 == 0:
        ## Plots
    ## Defini la taille de la figure et la taille de la police    
    plt.figure(figsize=(12,7))
    plt.rc('font', family='serif')
    plt.rc('font', size=20)
    
    ## Plot des trois trajectoires (passé, predite et vraie)
    plt.plot(a_past['x'],a_past['y'],marker='+',label = 'Past Trajectory')
    plt.plot(a_pred['x'],a_pred['y'], marker='x', label = 'Ground Truth')
    plt.plot(x_lin,y_lin,marker='o', label = 'Linear Prediction')
    plt.legend()
    plt.axis('equal')
    #
        #plt.savefig('./fig_stanf%s'%i, bbox_inches='tight')
```
